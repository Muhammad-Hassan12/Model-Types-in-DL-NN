## 1. Loading Dataset and library

```python
from keras.datasets import imdb
```

## 2. Unpacking

```python
(train_data, train_lables), (test_data, test_lables) = imdb.load_data(num_words = 10000)
```

## 3. Vectorization

```python
import numpy as np

def vector_sequences(sequences, dimension = 10000):       # Since the largest value of characters in 10000(almost) so the dimension will be 10000
    results = np.zeros((len(sequences), dimension))       # now converting the result into 0s and 1s
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1
    return results
```

```python
x_train = vector_sequences(train_data)   
x_test = vector_sequences(test_data)
```

```python
y_train = vector_sequences(train_data)   
y_test = vector_sequences(test_data)
```

* Now the data looks like this!

```python
x_train[0]
```

## 4. Network Creation

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape = (10000, )))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))   # Since its a binary classification so that the activation function will be "sigmoid"
```

```python
model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics= ['accuracy'])
```

```python
x_val = x_train[:10000]
partial_x_train = x_train[10000:]
```

```python
y_val = y_train[:10000]
partial_y_train = y_train[10000:]
```

```python
print('y_val shape:', y_val.shape)
print('partial_y_train shape:', partial_y_train.shape)
```

```python
if y_val.ndim == 2 and y_val.shape[1] > 1:
    y_val = np.argmax(y_val, axis=1)
    y_val = y_val.reshape(-1, 1)

if partial_y_train.ndim == 2 and partial_y_train.shape[1] > 1:
    partial_y_train = np.argmax(partial_y_train, axis=1)
    partial_y_train = partial_y_train.reshape(-1, 1)
```

```python
print('Converted y_val shape:', y_val.shape)
print('Converted partial_y_train shape:', partial_y_train.shape)
```

```python
history = model.fit(partial_x_train, partial_y_train, epochs= 20, batch_size= 512, validation_data=(x_val, y_val))
```

## 5. Visual representation of training

```python
import matplotlib.pyplot as plt
```

```python
history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, len(loss_values)+1)
plt.plot(epochs, loss_values, 'bo', label = 'Training Loss')
plt.plot(epochs, val_loss_values, 'b', label = 'Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show
```

```python
plt.clf()
acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']
plt.plot(epochs, acc_values, 'bo', label = 'Training Acc')
plt.plot(epochs, val_acc_values, 'b', label = 'Validation Acc')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```
